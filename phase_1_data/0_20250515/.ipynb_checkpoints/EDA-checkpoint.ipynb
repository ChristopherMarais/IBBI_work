```python
import os
import json
import pandas as pd
import glob
from concurrent.futures import ProcessPoolExecutor
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
from matplotlib.patches import Patch
```

```python
# --- Configuration ---
# IMPORTANT: Modify this to the path of your directory containing the JSON files
json_directory = './data/'
# IMPORTANT: Modify this to your desired output CSV file name and path
output_csv_file = './image_metadata.csv'
# Number of parallel processes to use. None will use os.cpu_count().
# Adjust if needed, e.g., max_workers=4
max_workers = None

# --- Helper Function to Process a Single JSON File ---
def process_json_file(file_path):
    """
    Reads a single JSON file and extracts relevant information.

    Args:
        file_path (str): The path to the JSON file.

    Returns:
        dict: A dictionary containing the extracted data, or None if an error occurs.
    """
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)

        image_name = data.get('imagePath')
        height = data.get('imageHeight')
        width = data.get('imageWidth')
        class_label = data.get('image_class_label') # As per your example JSON
        shapes = data.get('shapes', []) # Default to empty list if 'shapes' key is missing
        num_shapes = len(shapes)

        return {
            'image_name': image_name,
            'height': height,
            'width': width,
            'class_label': class_label,
            'num_shapes': num_shapes,
            'shapes_object': shapes # Storing the shapes list as an object
        }
    except json.JSONDecodeError:
        print(f"Error decoding JSON from file: {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred while processing file {file_path}: {e}")
        return None

# --- Main Processing Logic ---
if __name__ == '__main__':
    # Find all JSON files in the directory
    json_files = glob.glob(os.path.join(json_directory, '*.json'))

    if not json_files:
        print(f"No JSON files found in directory: {json_directory}")
    else:
        print(f"Found {len(json_files)} JSON files to process.")

        all_data = []

        # Using ProcessPoolExecutor for parallel processing
        # This creates a pool of processes to execute calls asynchronously.
        # 'max_workers=None' will typically default to the number of processors on the machine.
        with ProcessPoolExecutor(max_workers=max_workers) as executor:
            # map applies the function to each item in the iterable (json_files)
            # and returns an iterator that yields the results in the order of calls.
            results = list(executor.map(process_json_file, json_files))

        # Filter out None results (from files that failed to process)
        all_data = [res for res in results if res is not None]

        if not all_data:
            print("No data was successfully extracted from the JSON files.")
        else:
            # Convert the list of dictionaries to a Pandas DataFrame
            df = pd.DataFrame(all_data)

            # Ensure desired column order (optional, but good for consistency)
            columns_order = ['image_name', 'height', 'width', 'class_label', 'num_shapes', 'shapes_object']
            df = df[columns_order]

            # Save the DataFrame to a CSV file
            try:
                df.to_csv(output_csv_file, index=False)
                print(f"DataFrame successfully saved to {output_csv_file}")
            except Exception as e:
                print(f"Error saving DataFrame to CSV: {e}")
```

```python
# minor cleaning of dataframe
df = pd.read_csv("./image_metadata.csv")

# genus and species labels
split_data = df['class_label'].str.split("_", n=1, expand=True)
df["genus"] = split_data[0]
df["species"] = split_data[1]

# repalce 0 with unknown 
df["genus"] = df["genus"].str.replace("0", "unknown")

# make all text lowercase for species
df["species"] = df["species"].str.lower()

# image size calculation
df['size'] = np.sqrt(df['height']**2 + df['width']**2)

df
```

```python
# --- Define the conditions for each set ---
cond_has_shapes = df['num_shapes'] > 0
cond_known_genus = (~df['genus'].isin(['Unknown', 'unknown'])) & df['genus'].notna()
cond_known_species = (~df['species'].isin(['Unknown', 'unknown'])) & df['species'].notna() & cond_known_genus

# --- Calculate the sizes of the 7 disjoint regions for the Venn diagram ---
count_100 = (cond_has_shapes & ~cond_known_genus).sum()
count_010 = (~cond_has_shapes & cond_known_genus & ~cond_known_species).sum()
count_110 = (cond_has_shapes & cond_known_genus & ~cond_known_species).sum()
count_001 = 0
count_101 = 0
count_011 = (~cond_has_shapes & cond_known_species).sum()
count_111 = (cond_has_shapes & cond_known_species).sum()

total_images = len(df)
sum_of_venn_segments = count_100 + count_010 + count_110 + count_001 + count_101 + count_011 + count_111
count_none_of_the_three = total_images - sum_of_venn_segments

segment_counts = {
    '100': count_100, '010': count_010, '110': count_110,
    '001': count_001, '101': count_101,
    '011': count_011, '111': count_111
}

segment_styles = {
    '100': {'color': 'skyblue', 'alpha': 0.7, 'label_base': "Localization Only"},
    '010': {'color': 'gold', 'alpha': 0.7, 'label_base': "Genus Only"},
    '110': {'color': 'mediumpurple', 'alpha': 0.7, 'label_base': "Localization & Genus"},
    '011': {'color': 'yellow', 'alpha': 0.6, 'label_base': "Genus & Species Only"},
    '111': {'color': 'lightgreen', 'alpha': 0.8, 'label_base': "All"}
}

# --- Create the Venn Diagram ---
plt.figure(figsize=(30, 30)) # Your specified figure size

v = venn3(
    subsets=(count_100, count_010, count_110, count_001, count_101, count_011, count_111),
    set_labels=('', '', ''),
    set_colors=('darkgray', 'darkgray', 'darkgray'),
    alpha=0.7
)

# --- Hide Main Set Labels ---
if v.set_labels:
    for label in v.set_labels:
        if label:
            label.set_visible(False)

# --- Style Segments, Hide Zero-Count Segments, Hide Segment Count Labels, Remove Borders ---
all_segment_ids = ['100', '010', '110', '001', '101', '011', '111']
for segment_id in all_segment_ids:
    current_count = segment_counts.get(segment_id, 0)
    patch = v.get_patch_by_id(segment_id)
    count_label_text = v.get_label_by_id(segment_id)

    if count_label_text:
        count_label_text.set_visible(False)

    if patch:
        if current_count == 0:
            patch.set_visible(False)
        else:
            patch.set_visible(True)
            patch.set_edgecolor('none')
            if segment_id in segment_styles:
                style = segment_styles[segment_id]
                patch.set_facecolor(style['color'])
                patch.set_alpha(style['alpha'])

# --- Create and Display Legend (including "None of these") ---
legend_handles_segments = []
for segment_id, style_info in segment_styles.items():
    count = segment_counts.get(segment_id, 0)
    if count > 0:
        legend_handles_segments.append(
            Patch(facecolor=style_info['color'],
                  edgecolor='grey',
                  alpha=style_info['alpha'],
                  label=f"{style_info['label_base']}: {count}")
        )

color_none_segment = 'white'
alpha_none_segment = 0.8
legend_handles_segments.append(
    Patch(facecolor=color_none_segment,
          edgecolor='grey',
          alpha=alpha_none_segment,
          label=f"None: {count_none_of_the_three}")
)
color_none_segment = 'grey'
alpha_none_segment = 0.8
legend_handles_segments.append(
    Patch(facecolor=color_none_segment,
          edgecolor='grey',
          alpha=alpha_none_segment,
          label=f"Total Images: {total_images}")
)

if legend_handles_segments:
    ax = plt.gca()
    leg = ax.legend(handles=legend_handles_segments,
                    loc='upper center',
                    bbox_to_anchor=(0.5, 0.0), # Centered below the diagram
                    fontsize=22,           # Legend text "a bit larger"
                    title="Metadata Coverage",
                    title_fontsize=24,   # Legend title "a bit larger"
                    handlelength=3,
                    labelspacing=1.0,
                    columnspacing=2.5,
                    ncol=3,
                    frameon=False)

    # Removed the section that made legend text/title bold
    # if leg:
    #     for text in leg.get_texts():
    #         text.set_fontweight('bold') # REMOVED
    #     leg.get_title().set_fontweight('bold') # REMOVED

# --- Title and Plot Adjustments ---
plt.title(f"\nImage Metadata Distribution",
          fontsize=24, # Kept from your last code
          pad=20,
          loc='center',
          fontweight='bold') # Main title remains bold

# Adjust subplot parameters to make space AT THE BOTTOM for the legend
# May need to increase bottom further if legend text is much larger
plt.subplots_adjust(left=0.05, right=0.95, top=0.90, bottom=0.5) # Increased bottom margin slightly
plt.savefig("Image_metadata_distribution.svg", bbox_inches='tight')
plt.show()

# --- Verification Print Statements ---
print(f"Total images: {total_images}")
print(f"Images with Localization (shapes): {cond_has_shapes.sum()}")
print(f"Images with known genus: {cond_known_genus.sum()}")
print(f"Images with known species (and known genus): {cond_known_species.sum()}")
print(f"\nNone of these (No Loc. & Unknown Genus): {count_none_of_the_three}")
print("\nSegment Counts for Legend (Non-Zero Styled Segments):")
for segment_id, style_info in segment_styles.items():
    count = segment_counts.get(segment_id, 0)
    if count > 0:
        print(f"  {style_info['label_base']}: {count}")
```

```python
# --- Create Genus Summary DataFrame (remains the same) ---
genus_summary_df = df.groupby('genus').agg(
    num_images=('num_shapes', 'size'), # 'size' counts rows in each group
    total_shapes=('num_shapes', 'sum')
).reset_index()

# --- Create a combined 'genus_species' column for unique species identification ---
# Using .astype(str) to handle potential None or numerical values in genus/species columns gracefully
df['full_species_name'] = df['genus'].astype(str) + "_" + df['species'].astype(str)

# --- Create Species Summary DataFrame using the 'full_species_name' ---
species_summary_df = df.groupby('full_species_name').agg(
    num_images=('num_shapes', 'size'),
    total_shapes=('num_shapes', 'sum')
).reset_index() # Moves 'full_species_name' from index to a column

# --- Save DataFrames to Disk ---
genus_summary_filename = 'genus_summary_stats.csv'
species_summary_filename = 'species_summary_stats.csv' # Filename remains the same, content changes

try:
    genus_summary_df.to_csv(genus_summary_filename, index=False)
    species_summary_df.to_csv(species_summary_filename, index=False) # This will now have 'full_species_name'
    print(f"Genus Summary DataFrame saved to '{genus_summary_filename}'")
    print(f"Species Summary DataFrame (with full names) saved to '{species_summary_filename}'")
except Exception as e:
    print(f"Error saving files: {e}")

print("\nFirst 5 rows of Genus Summary:")
print(genus_summary_df.head())
print("\nFirst 5 rows of Species Summary (with 'full_species_name'):")
print(species_summary_df.head())
```

```python
# --- Load the summary DataFrames ---
genus_summary_filename = 'genus_summary_stats.csv'
species_summary_filename = 'species_summary_stats.csv' # This should have 'full_species_name'

try:
    genus_summary_df = pd.read_csv(genus_summary_filename)
    species_summary_df = pd.read_csv(species_summary_filename)
    print("Summary DataFrames loaded successfully.")
    print(f"Initial species_summary_df head:\n{species_summary_df.head()}")
    print(f"Initial genus_summary_df head:\n{genus_summary_df.head()}")
except FileNotFoundError:
    print(f"ERROR: Summary files ('{genus_summary_filename}', '{species_summary_filename}') not found.")
    print("Please ensure Cell 1 has been run successfully to create these files.")
    # Create empty DataFrames with expected columns to prevent downstream errors
    genus_summary_df = pd.DataFrame(columns=['genus', 'num_images', 'total_shapes'])
    species_summary_df = pd.DataFrame(columns=['full_species_name', 'num_images', 'total_shapes'])


# --- Filter out entries with "Unknown" in their names ---
if not genus_summary_df.empty:
    # Ensure 'genus' column is treated as string for filtering
    original_genus_count = len(genus_summary_df)
    genus_summary_df = genus_summary_df[
        ~genus_summary_df['genus'].astype(str).str.contains('unknown', case=False, na=False)
    ]
    print(f"\nFiltered genus_summary_df: {original_genus_count - len(genus_summary_df)} entries with 'unknown' in genus name removed. New shape: {genus_summary_df.shape}")
    print(f"Filtered genus_summary_df head:\n{genus_summary_df.head()}")


if not species_summary_df.empty:
    # The column containing combined species names should be 'full_species_name'
    # as created in the modified Cell 1
    species_name_column = 'full_species_name' 
    
    if species_name_column not in species_summary_df.columns:
        # Fallback check if the column was just 'species' but contained combined names
        if 'species' in species_summary_df.columns:
            print(f"Warning: Column '{species_name_column}' not found. Using 'species' column for filtering species names.")
            species_name_column = 'species'
        else:
            print(f"Warning: Expected species name column ('{species_name_column}' or 'species') not found in species_summary_df. Skipping species name filtering.")
            species_name_column = None # Will skip filtering if no valid column

    if species_name_column and species_name_column in species_summary_df.columns:
        original_species_count = len(species_summary_df)
        species_summary_df = species_summary_df[
            ~species_summary_df[species_name_column].astype(str).str.contains('unknown', case=False, na=False)
        ]
        print(f"\nFiltered species_summary_df: {original_species_count - len(species_summary_df)} entries with 'unknown' in '{species_name_column}' removed. New shape: {species_summary_df.shape}")
        print(f"Filtered species_summary_df head:\n{species_summary_df.head()}")


# --- Helper Function for Plotting Rank-Frequency (remains the same as your last version) ---
def plot_rank_frequency(df, value_col, name_col_for_legend, entity_display_name, value_display_name, line_color):
    """
    Generates a rank-frequency plot with a legend for top 5 entities.
    """
    if df.empty or value_col not in df.columns or name_col_for_legend not in df.columns:
        print(f"DataFrame for {entity_display_name} is empty or missing required columns ('{value_col}', '{name_col_for_legend}'). Skipping plot: '{value_display_name} per {entity_display_name}'")
        return

    df_sorted = df.sort_values(by=value_col, ascending=False).reset_index(drop=True)

    if df_sorted.empty:
        print(f"No data to plot for {entity_display_name} after sorting (all values might be NaN or list was empty). Skipping plot.")
        return
        
    plt.figure(figsize=(14, 8))
    plt.plot(df_sorted.index, df_sorted[value_col], marker='.', linestyle='-', color=line_color)
    
    title = f"{value_display_name} per {entity_display_name}" # Updated title
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel(f"{entity_display_name} (Rank Order)", fontsize=14) # Updated x-label
    plt.ylabel(value_display_name, fontsize=14)
    
    plt.xticks([])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    top_5_df = df_sorted.head(10)
    legend_handles = []
    if not top_5_df.empty:
        for _idx, row_data in top_5_df.iterrows():
            entity_label_name = str(row_data[name_col_for_legend])
            if len(entity_label_name) > 35:
                entity_label_name = entity_label_name[:32] + "..."
            
            label_text = f"{entity_label_name}: {int(row_data[value_col])}"
            legend_handles.append(Patch(facecolor=line_color, alpha=0.7, label=label_text))
            
        plt.legend(handles=legend_handles, 
                   title=f"Top 10 {entity_display_name}", 
                   loc='upper right', 
                   fontsize=10, 
                   title_fontsize=12,
                   frameon=True)

    plt.tight_layout()
    plt.savefig(f"{title}.svg", bbox_inches='tight')
    plt.show()

# --- Define a list of distinct colors for the plots ---
plot_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# --- Generate the 4 Plots using the filtered DataFrames ---

# Plot 1: Number of images per Species
plot_rank_frequency(species_summary_df, 
                    value_col='num_images', 
                    name_col_for_legend='full_species_name', # Should match your CSV column from Cell 1
                    entity_display_name='Species', 
                    value_display_name='Number of Images', 
                    line_color=plot_colors[0])

# Plot 2: Number of images per Genus
plot_rank_frequency(genus_summary_df, 
                    value_col='num_images', 
                    name_col_for_legend='genus', 
                    entity_display_name='Genus', 
                    value_display_name='Number of Images', 
                    line_color=plot_colors[1])

# Plot 3: Total num_shapes per Species
plot_rank_frequency(species_summary_df, 
                    value_col='total_shapes', 
                    name_col_for_legend='full_species_name', # Should match your CSV column
                    entity_display_name='Species', 
                    value_display_name='Total Number of Localizations', 
                    line_color=plot_colors[2])

# Plot 4: Total num_shapes per Genus
plot_rank_frequency(genus_summary_df, 
                    value_col='total_shapes', 
                    name_col_for_legend='genus', 
                    entity_display_name='Genus', 
                    value_display_name='Total Number of Localizations', 
                    line_color=plot_colors[3])
```

```python
# --- Load the summary DataFrames ---
genus_summary_filename = 'genus_summary_stats.csv'
species_summary_filename = 'species_summary_stats.csv' # This should have 'full_species_name'

try:
    genus_summary_df = pd.read_csv(genus_summary_filename)
    species_summary_df = pd.read_csv(species_summary_filename)
    print("Summary DataFrames loaded successfully.")
except FileNotFoundError:
    print(f"ERROR: Summary files ('{genus_summary_filename}', '{species_summary_filename}') not found.")
    print("Please ensure Cell 1 has been run successfully to create these files.")
    genus_summary_df = pd.DataFrame(columns=['genus', 'num_images', 'total_shapes'])
    species_summary_df = pd.DataFrame(columns=['full_species_name', 'num_images', 'total_shapes'])

# --- Filter out entries with "Unknown" in their names (from your previous request) ---
if not genus_summary_df.empty:
    original_genus_count = len(genus_summary_df)
    genus_summary_df = genus_summary_df[
        ~genus_summary_df['genus'].astype(str).str.contains('unknown', case=False, na=False)
    ]
    print(f"\nFiltered genus_summary_df: {original_genus_count - len(genus_summary_df)} entries with 'unknown' in genus name removed.")

if not species_summary_df.empty:
    species_name_column = 'full_species_name'
    if species_name_column not in species_summary_df.columns and 'species' in species_summary_df.columns:
        species_name_column = 'species'
    
    if species_name_column in species_summary_df.columns:
        original_species_count = len(species_summary_df)
        species_summary_df = species_summary_df[
            ~species_summary_df[species_name_column].astype(str).str.contains('unknown', case=False, na=False)
        ]
        print(f"Filtered species_summary_df: {original_species_count - len(species_summary_df)} entries with 'unknown' in '{species_name_column}' removed.")

# --- Helper Function for Plotting Combined Rank-Frequency ---
def plot_combined_rank_frequency(df_summary, 
                                 value_col1, value_col1_display, color1,
                                 value_col2, value_col2_display, color2,
                                 entity_display_name):
    """
    Generates a combined rank-frequency plot for two metrics on the same figure,
    using a primary and a secondary y-axis.
    Each metric is sorted independently for its rank order.
    """
    if df_summary.empty:
        print(f"DataFrame for {entity_display_name} is empty. Skipping combined plot.")
        return
    if not all(col in df_summary.columns for col in [value_col1, value_col2]):
        print(f"DataFrame for {entity_display_name} is missing one or more value columns ('{value_col1}', '{value_col2}'). Skipping combined plot.")
        return

    # Sort independently for each metric
    df_sorted1 = df_summary.sort_values(by=value_col1, ascending=False).reset_index(drop=True)
    df_sorted2 = df_summary.sort_values(by=value_col2, ascending=False).reset_index(drop=True)

    if df_sorted1.empty and df_sorted2.empty:
        print(f"No data to plot for {entity_display_name} after sorting. Skipping combined plot.")
        return

    fig, ax1 = plt.subplots(figsize=(14, 8))

    # Plot first metric on ax1
    if not df_sorted1.empty:
        line1 = ax1.plot(df_sorted1.index, df_sorted1[value_col1], marker='.', linestyle='-', color=color1, label=value_col1_display)
    else:
        line1 = [] # Empty list if no data for line1
        print(f"No data for '{value_col1_display}' for {entity_display_name}.")
        
    ax1.set_xlabel(f"{entity_display_name} (Rank Order - Individual names not shown, 'Unknown' Excluded)", fontsize=14)
    ax1.set_ylabel(value_col1_display, fontsize=14, color=color1 if line1 else 'black')
    if line1: ax1.tick_params(axis='y', labelcolor=color1)
    ax1.set_xticks([]) # Hide x-axis numerical ticks (rank is implicit)
    ax1.grid(axis='y', linestyle='--', alpha=0.7, which='major', color=color1 if line1 else 'grey')

    # Create a secondary y-axis for the second metric
    ax2 = ax1.twinx()
    if not df_sorted2.empty:
        line2 = ax2.plot(df_sorted2.index, df_sorted2[value_col2], marker='x', linestyle='--', color=color2, label=value_col2_display)
    else:
        line2 = [] # Empty list if no data for line2
        print(f"No data for '{value_col2_display}' for {entity_display_name}.")

    ax2.set_ylabel(value_col2_display, fontsize=14, color=color2 if line2 else 'black')
    if line2: ax2.tick_params(axis='y', labelcolor=color2)
    # ax2 does not need its own grid or x-label as it shares with ax1

    plt.title(f"Distribution of {value_col1_display} & {value_col2_display} per {entity_display_name}", fontsize=16, fontweight='bold')
    
    # Add legend for the two lines
    # Collect lines that were actually plotted
    plotted_lines = [l[0] for l in [line1, line2] if l] # Get the Line2D objects if line1/line2 are not empty
    if plotted_lines:
        labels = [l.get_label() for l in plotted_lines]
        ax1.legend(plotted_lines, labels, loc='upper right', fontsize=10, title="Metrics", title_fontsize=12, frameon=True)
    
    fig.tight_layout() # Adjust plot to ensure everything fits
    plt.show()

# --- Define colors for the metrics ---
color_num_images = '#1f77b4'  # Blue
color_total_shapes = '#ff7f0e' # Orange

# --- Generate the 2 Combined Plots ---

# Figure 1: Combined plot for Species
print("\n--- Generating Combined Plot for Species ---")
plot_combined_rank_frequency(df_summary=species_summary_df,
                             value_col1='num_images', value_col1_display='Number of Images', color1=color_num_images,
                             value_col2='total_shapes', value_col2_display='Total Number of Shapes', color2=color_total_shapes,
                             entity_display_name='Species')

# Figure 2: Combined plot for Genus
print("\n--- Generating Combined Plot for Genus ---")
plot_combined_rank_frequency(df_summary=genus_summary_df,
                             value_col1='num_images', value_col1_display='Number of Images', color1=color_num_images,
                             value_col2='total_shapes', value_col2_display='Total Number of Shapes', color2=color_total_shapes,
                             entity_display_name='Genus')
```
